```python
from keras.preprocessing.image import img_to_array, load_img
from keras import backend as K

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import os
import cv2
from tensorflow import keras
model = keras.models.load_model("model_mnist.h5")
```

```python
dir_path = r"F:\Mnist Digit Classification\handwritten_numbers" #image dir
```

```python
for i in os.listdir(dir_path):    
    img = load_img(dir_path+"/"+i, grayscale=True)
    plt.imshow(img,cmap='gray')
    plt.show()
    x=img.resize((28,28)) #resize images
    x = img_to_array(x)
    x = np.expand_dims(x, axis=0) 
    y_pred = model.predict(x)
    print("The predicted number is: ")
    print("\n")
    print((y_pred.argmax()))
    print("\n")

```
